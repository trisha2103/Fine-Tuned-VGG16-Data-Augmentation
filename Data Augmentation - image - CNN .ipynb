# Import Req packages

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

# Modify the image using ImageDataGenerator

```python
gen = ImageDataGenerator(
    rotation_range=10, 
    width_shift_range=0.1, 
    height_shift_range=0.1, 
    shear_range=0.15, 
    zoom_range=0.1, 
    channel_shift_range=10., 
    zca_whitening=True,
    horizontal_flip=True)
```

```python
choose_img = random.choice(os.listdir('C:/Users/yuvar/Deep learning/dogs-vs-cats/train/dog'))
```

```python
img_path = 'C:/Users/yuvar/Deep learning/dogs-vs-cats/train/dog/' + choose_img
```

```python
image = np.expand_dims(plt.imread(img_path),0)
```

```python
plt.imshow(image[0])
```

# Generate batches of augmented images from the original image

```python
aug_iter = gen.flow(image)
```

```python
aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
plotImages(aug_images)
```
